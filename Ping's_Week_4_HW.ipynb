# Creating and Customizing Bar Charts with Matplotlib

```python
from matplotlib import pyplot as plt
import numpy as np
```

### Task 1: Bar chart of 2023 crime rates

```python
# Question: How can I makes the small values (like 5.7) more visible without losing the other higher rates' visuals?
# Data
crimes = ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery']
rates = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]

# Bar chart
plt.figure(figsize=(10, 6))
plt.bar(crimes, rates, color= 'pink')

# Labels, title, grid
plt.title('Offenses per 100,000 People in 2023', fontsize=16, fontweight='bold', pad=12)
plt.xlabel('Offense Type', fontsize=12)
plt.ylabel('Offenses per 100,000 People', fontsize=12)
plt.axhline(0, color='black', linewidth=0.8)
plt.xticks(fontsize=9)
plt.grid(axis='y', linestyle='--', color='grey', alpha=0.7)

plt.tight_layout()
plt.show()
```

### Task 2: Horizontal bar chart of % changes (2022–2023)

```python
crimes = ['Murder', 'Rape', 'Robbery', 'Aggravated Assault', 'Burglary', 'Motor Vehicle Theft', 'Larceny']
changes_in_percentage = [-12.0, -9.8, -0.8, -3.3, -8.1, +12.0, -4.9]

colors = ['tomato', 'orange', 'gold', 'lightgreen', 'skyblue', 'plum', 'gray']
# bar chart
plt.figure(figsize=(12, 8))
plt.barh(crimes, changes_in_percentage, color=colors, height=0.7)

# labels
plt.title('22-23 Crime Rate Changes In Percentage')
plt.xlabel('Amount As In Millions')
plt.ylabel('Crime Category')

# grid and horizontal line
plt.axhline(0, color='black', linewidth=0.9)
plt.grid(axis='y', linestyle='solid', color='grey', alpha=0.9)
plt.grid(axis='x', linestyle='solid', color='grey', alpha=0.9)

plt.tight_layout()
plt.show()
```

```python
# Version 3. I think this one is the best bc you can see the decrease and increase immediate
# Question:
# I gave different colors to the treant: increase = orange, decrease = green, how to label in legend to show green is decrease and orange is increase?
# How to add data labels on each bar?
# Why does the last bar always overlap the line in all three versions? I tried to adjust it but didn't work out.

crimes = ['Murder', 'Rape', 'Robbery', 'Aggravated Assault',
          'Burglary', 'Motor Vehicle Theft', 'Larceny']
changes_in_percentage = [-12.0, -9.8, -0.8, -3.3, -8.1, +12.0, -4.9]


colors = ['tomato' if value > 0 else 'teal' for value in changes_in_percentage]

plt.figure(figsize=(13, 7))
plt.barh(crimes, changes_in_percentage, color=colors, height=0.6)


plt.title('Trend in U.S. Crime Rates (2022–2023)', fontsize=16, fontweight='bold')
plt.xlabel('Change in Crime Rate (%)', fontsize=12)
plt.ylabel('Crime Type', fontsize=12)

plt.axvline(0, color='black', linewidth=0.9)
plt.grid(axis='y', linestyle='solid', color='grey', alpha=0.6)
plt.grid(axis='x', linestyle='solid', color='grey', alpha=0.6)

plt.tight_layout()
plt.show()
```

Task 3: Multi-Series (Grouped) Bar Chart with fake data (Company A–C, Products 1–3 → 3 bars per company)

```python
# Fake sales data of hockey gear companies
# Products = ['Helmet', 'Sticks', 'Skates']
# Bauer [20, 35, 30]
# CCM  [25, 32, 34]
# Warrior [22, 30, 29]

x = ['Bauer', 'CCM', 'Warrior']
y = [20, 35, 30]
z = [25, 32, 34]
i = [22, 30, 29]

x_indexes = np.arange(len(x))

fig, ax = plt.subplots(figsize=(8, 6))

print(x_indexes)

plt.bar(x_indexes - 0.25, y, width=0.25, label='Helmet', color = 'pink' )
plt.bar(x_indexes, z, width=0.25, label='Stick', color = 'Seagreen' )
plt.bar(x_indexes + 0.25, i, width=0.25, label='Skates', color = 'cornflowerblue' )

plt.title('Hockey Gear Sales by Company and Product')
plt.ylabel('Sales in Million')
plt.xticks(ticks=x_indexes, labels=x)
plt.legend(title='Products', loc='upper left', fontsize=8)

plt.tight_layout()
plt.show()
```
